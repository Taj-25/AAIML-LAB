```python
import numpy as np
import pandas as pd
#from sklearn.preprocessing import StandardScaler
#from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Load feature names
features = [f"{feat}_{i}" for i, feat in enumerate(pd.read_csv('features.txt', delim_whitespace=True, header=None)[1])]
```

```python
features
```

```python

# Load training and test data
X_train = pd.read_csv('X_train.txt', delim_whitespace=True, header=None, names=features)
X_test = pd.read_csv('X_test.txt', delim_whitespace=True, header=None, names=features)
y_train = pd.read_csv('y_train.txt', delim_whitespace=True, header=None)[0]
y_test = pd.read_csv('y_test.txt', delim_whitespace=True, header=None)[0]
print(X_train.head())  
```

```python
# Train KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print("KNN Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
# Train SVC classifier
svc = SVC(kernel='rbf', C=1.0)
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
print("SVC Accuracy:", accuracy_score(y_test, y_pred_svc))
print("SVC Classification Report:\n", classification_report(y_test, y_pred_svc))
```
